# Amazon Data Visualization

## 1. Imports and Initial Data

```python
import pandas as pd 
import numpy as np 
import plotly.express as px 
import matplotlib.pyplot as plt

df = pd.read_csv('amazon.csv')
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df['discount_percentage'] = df['discount_percentage'].str.rstrip('%').astype('float') / 100.0

df['discounted_price']=df['discounted_price'].str.replace(',','')
df['actual_price']=df['actual_price'].str.replace(',','')
df['rating_count']=df['rating_count'].str.replace(',','')

df['discounted_price'] = df['discounted_price'].str.lstrip('₹').astype('float') 
df['actual_price'] = df['actual_price'].str.lstrip('₹').astype('float') 
df['rating_count'] = df['rating_count'].fillna(0)
df['rating_count'] = df['rating_count'].astype('int')

df.loc[df['rating'] == '|']
df.drop(index=1279 , inplace= True)

df['rating'] = df['rating'].astype('float')
```

```python
df
```

```python
categories = []

for c in df['category']:
    categories += c.split('|')

unique_category = list(set(categories))

unique_category
```

```python
fig = px.scatter(df, x='rating_count', y='actual_price')
fig.show()
```

```python
fig = px.imshow(df.corr())
fig.show()
```

```python
fig = px.bar(df['rating'].value_counts(), title='Value Counts for Each Rating')

fig.update_layout(
    showlegend=False
)

fig.show()
```

```python
fig = px.scatter(df, x='discount_percentage', y='rating_count', color='rating', title='Relationship Between Discount, Number of Ratings, and Rating')
fig.show()
```
